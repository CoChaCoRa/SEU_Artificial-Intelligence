```python
import random
import numpy as np
import time
import os
import json
import copy

# 启发式评估函数
# param: 当前棋盘布局状态
# return: 当前布局存在冲突的皇后对数，即评估函数值h
def get_numof_conflict(status):
    num = 0
    for i in range(len(status)):
        for j in range(i + 1,len(status)):
            # 同列：status[i] == status[j]
            if status[i] == status[j]:
                num += 1
            # 对角线：abs(status[i]-status[j]) == j - i
            offset = j - i
            if abs(status[i]-status[j]) == offset:
                num += 1
    return num

# 爬山法
# param： 当前棋盘布局状态
# return： 最佳后继的布局状态
def hill_climbing(status):
    convert = {}    # 存储所有可能的后继状态，Key是所有可能后继状态，Value是此状态对应的评估函数值
    length = len(status)

    for col in range(length):
        best_move = status[col]
        for row in range(length):
            if status[col] == row:
                continue
            status_copy = list(status)
            status_copy[col] = row
            convert[(col,row)] = get_numof_conflict(status_copy)
 
    answers = [] # 存储所有的最佳后继集合
    conflict_now = get_numof_conflict(status) # 当前状态评估函数值
 
    # 遍历存储所有可能后继的字典，找出最佳后继
    for key,value in convert.items():
        if value < conflict_now:
            conflict_now = value
    for key,value in convert.items():
        if value == conflict_now:
            answers.append(key)
 
    # 如果最佳后继集合元素大于一个，随机选择一个
    if len(answers) > 0:
        x = random.randint(0,len(answers)-1)
        col = answers[x][0]
        row = answers[x][1]
        status[col] = row
 
    return status
 
# 求得满足冲突数为0的一个解
def Queens(size, filename):
    
    closelist = [] # 存放所有扩展过的节点
    
    # 初始状态：所有皇后都在对角线上
    status = []
    for i in range(size):
        status.append(i)
    print("棋盘初始状态：")
    print(status,'\n')
    print("棋盘初始冲突数：")
    print(get_numof_conflict(status),'\n')
    
    time_start = time.time()
    # 当存在冲突的个数大于0时循环求解最佳后继，直到找到解
    step = 0
    while get_numof_conflict(status) > 0:
        status = hill_climbing(status)
        node = {}
        node['status'] = copy.deepcopy(status)
        node['value'] = get_numof_conflict(status)
        closelist.append(node)
        print(status)
        # print(closelist[cnt]['status'])
        print(get_numof_conflict(status),'\n')
        step += 1
    #print(closelist)
        
    time_end = time.time()
    time_cost = time_end - time_start
    
    print("棋盘最终状态：")
    print(status)
    
    # 将结果写入文件并输出
    h = open(filename, 'w', encoding='utf-8')
    h.write('问题规模：' + str(size) + '\n')
    h.write('搜索时间：' + str(time_cost) + '\n')
    h.write('搜索步数：' + str(step) + '\n')
    for node in closelist:
        h.write('状态：' + str(node['status']) + '\n')
        h.write('评估函数值：' + str(node['value']) + '\n')
    
    board = [0] * len(status)
    for ix in range(len(status)):
        board[ix] = [0] * len(status)
        
    for i in range(len(status)):
        for j in range(len(status)):
            if status[i] == j:
                board[i][j] = 1
            j += 1
        i += 1
    temp = []
    for row in board:
        temp.append(row)
    print(str(np.array(temp)) + '\n')
    h.write(str(np.array(temp)) + '\n')
    h.close()
```
